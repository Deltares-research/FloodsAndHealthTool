```python
import os 
from osgeo import gdal
import numpy as np
```

```python
# Input folder containing raster files
input_folder = 'C:/Users/bleser/OneDrive - Stichting Deltares/Desktop/Floods&Health Quantification/Data bronnen/tif_raw_moz/Children_raw'

# Output raster file
output_file = 'Children_test_output_raster.tif'

# Output folder for the raster file
output_folder = 'C:/Users/bleser/OneDrive - Stichting Deltares/Desktop/Floods&Health Quantification/Data bronnen/tif_raw_moz/Children_raw'

# Get a list of all raster files in the input folder
raster_files = [f for f in os.listdir(input_folder) if f.endswith('.tif')]

if len(raster_files) == 0:
    raise Exception("No raster files found in the input folder.")

# Open the first raster file to get the desired extent and properties
first_raster_path = os.path.join(input_folder, raster_files[0])

try:
    first_raster = gdal.Open(first_raster_path)
    if first_raster is None:
        raise Exception(f"Failed to open raster file: {first_raster_path}")
    
    driver = gdal.GetDriverByName('GTiff')
    rows, cols, bands = first_raster.RasterYSize, first_raster.RasterXSize, first_raster.RasterCount
    geotransform = first_raster.GetGeoTransform()
    projection = first_raster.GetProjection()
    
except Exception as e:
    raise Exception(f"Error opening raster file: {first_raster_path}\n{str(e)}")

# Create an array to accumulate the sum of raster values
sum_array = np.zeros((rows, cols), dtype=np.float32)

# Loop through each input raster file
for file_name in raster_files:
    file_path = os.path.join(input_folder, file_name)
    
    try:
        input_raster = gdal.Open(file_path)
        if input_raster is None:
            raise Exception(f"Failed to open raster file: {file_path}")
        
        # Read the raster data
        raster_data = input_raster.ReadAsArray()
        
        # Accumulate the sum of raster values
        sum_array += raster_data.astype(np.float32)
        
    except Exception as e:
        raise Exception(f"Error processing raster file: {file_path}\n{str(e)}")
        
    finally:
        input_raster = None  # Close the input raster file

# Create the output raster file
output_path = os.path.join(input_folder, output_file)

try:
    output_raster = driver.Create(output_path, cols, rows, bands, gdal.GDT_Float32)
    output_raster.SetGeoTransform(geotransform)
    output_raster.SetProjection(projection)
    output_raster.GetRasterBand(1).WriteArray(sum_array)
    
except Exception as e:
    raise Exception(f"Error creating output raster file: {output_path}\n{str(e)}")

finally:
    output_raster = None  # Close the output raster file

print("Output raster file saved to:", output_path)
```

```python
# Input folder containing raster files for adults
input_folder_adults = 'C:/Users/bleser/OneDrive - Stichting Deltares/Desktop/Floods&Health Quantification/Data bronnen/tif_raw_moz/Adults_raw'

# Output folder
output_folder = 'C:/Users/bleser/OneDrive - Stichting Deltares/Desktop/Floods&Health Quantification/Data bronnen/tif_raw_moz/Output'

# Output raster file for adults
output_file_adults = 'combined_output_raster_adults.tif'

# Get a list of all raster files in the input folder for adults
raster_files_adults = [f for f in os.listdir(input_folder_adults) if f.endswith('.tif')]

if len(raster_files_adults) == 0:
    raise Exception("No raster files found in the input folder for adults.")

# Open the first raster file to get the desired extent and properties
first_raster_path_adults = os.path.join(input_folder_adults, raster_files_adults[0])
first_raster_adults = gdal.Open(first_raster_path_adults)

if first_raster_adults is None:
    raise Exception(f"Failed to open raster file: {first_raster_path_adults}")

driver = gdal.GetDriverByName('GTiff')
rows, cols, bands = first_raster_adults.RasterYSize, first_raster_adults.RasterXSize, first_raster_adults.RasterCount
geotransform = first_raster_adults.GetGeoTransform()
projection = first_raster_adults.GetProjection()

# Create an array to accumulate the combined raster data
combined_data_adults = np.zeros((rows, cols), dtype=np.float32)

# Loop through each input raster file
for file_name in raster_files_adults:
    file_path_adults = os.path.join(input_folder_adults, file_name)
    input_raster_adults = gdal.Open(file_path_adults)
    
    # Read the raster data
    raster_data_adults = input_raster_adults.ReadAsArray()
    
    # Set negative values to zero
    raster_data_adults[raster_data_adults < 0] = 0
    
    # Accumulate the raster data
    combined_data_adults += raster_data_adults.astype(np.float32)
    
    input_raster_adults = None  # Close the input raster file

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Create the output raster file
output_path = os.path.join(output_folder, output_file_adults)
output_raster_adults = driver.Create(output_path, cols, rows, bands, gdal.GDT_Float32)
output_raster_adults.SetGeoTransform(geotransform)
output_raster_adults.SetProjection(projection)

# Write the combined data to the output file
output_raster_adults.GetRasterBand(1).WriteArray(combined_data_adults)

# Close the output raster file and save it
output_raster_adults = None

print("Combined raster file saved to:", output_path)
```

```python
import os
import numpy as np
from osgeo import gdal

# Input folders containing raster files for children and adults
input_folder_children = r'C:\Users\bleser\OneDrive - Stichting Deltares\Desktop\Floods&Health Quantification\Data bronnen\tif_raw_moz\Output\combined_output_raster_children.tif'
input_folder_adults = r'C:\Users\bleser\OneDrive - Stichting Deltares\Desktop\Floods&Health Quantification\Data bronnen\tif_raw_moz\Output\combined_output_raster_adults.tif'

# Output folder
output_folder = r'C:\Users\bleser\OneDrive - Stichting Deltares\Desktop\Floods&Health Quantification\Data bronnen\tif_raw_moz\Output'

# Output raster file
output_file = 'children_and_adults_combined_seperate.tif'

# Get a list of raster files for children and adults
raster_files_children = [f for f in os.listdir(input_folder_children) if f.endswith('.tif')]
raster_files_adults = [f for f in os.listdir(input_folder_adults) if f.endswith('.tif')]

if len(raster_files_children) == 0 or len(raster_files_adults) == 0:
    raise Exception("No raster files found in the input folders.")

# Open the first raster file to get information for the output file
first_raster_path = os.path.join(input_folder_children, raster_files_children[0])
first_raster = gdal.Open(first_raster_path)

if first_raster is None:
    raise Exception(f"Failed to open raster file: {first_raster_path}")

driver = gdal.GetDriverByName('GTiff')
rows, cols, bands = first_raster.RasterYSize, first_raster.RasterXSize, 2  # Two bands for children and adults
geotransform = first_raster.GetGeoTransform()
projection = first_raster.GetProjection()

# Create arrays to accumulate the combined raster data for children and adults
combined_data_children = np.zeros((rows, cols), dtype=np.float32)
combined_data_adults = np.zeros((rows, cols), dtype=np.float32)

# Loop through each input raster file for children
for file_name in raster_files_children:
    file_path = os.path.join(input_folder_children, file_name)
    input_raster = gdal.Open(file_path)
    
    # Read the raster data
    raster_data = input_raster.ReadAsArray()
    
    # Set negative values to zero
    raster_data[raster_data < 0] = 0
    
    # Accumulate the raster data for children
    combined_data_children += raster_data.astype(np.float32)
    
    input_raster = None  # Close the input raster file

# Loop through each input raster file for adults
for file_name in raster_files_adults:
    file_path = os.path.join(input_folder_adults, file_name)
    input_raster = gdal.Open(file_path)
    
    # Read the raster data
    raster_data = input_raster.ReadAsArray()
    
    # Set negative values to zero
    raster_data[raster_data < 0] = 0
    
    # Accumulate the raster data for adults
    combined_data_adults += raster_data.astype(np.float32)
    
    input_raster = None  # Close the input raster file

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Create the output raster file
output_path = os.path.join(output_folder, output_file)
output_raster = driver.Create(output_path, cols, rows, bands, gdal.GDT_Float32)
output_raster.SetGeoTransform(geotransform)
output_raster.SetProjection(projection)

# Write the combined data to the output file
output_raster.GetRasterBand(1).WriteArray(combined_data_children)
output_raster.GetRasterBand(2).WriteArray(combined_data_adults)

# Close the output raster file and save it
output_raster = None

print("Combined raster file saved to:", output_path)
```

```python
import rasterio
from rasterio.enums import Resampling
import numpy as np

# Paths to input GeoTIFF files
input_files = [
    r'C:\Users\bleser\OneDrive - Stichting Deltares\Desktop\Floods&Health Quantification\Data bronnen\tif_raw_moz\moz_pd_2020.tif',
    r'C:\Users\bleser\OneDrive - Stichting Deltares\Desktop\Floods&Health Quantification\Data bronnen\tif_raw_moz\Output\combined_output_raster_adults.tif',
    r'C:\Users\bleser\OneDrive - Stichting Deltares\Desktop\Floods&Health Quantification\Data bronnen\tif_raw_moz\Output\combined_output_raster_children.tif'
]

# Band names corresponding to each input file
band_names = [
    'Population Density',
    'Adults',
    'Children'
]

# Read the datasets and find the smallest raster's shape
smallest_raster_shape = None
data_type = None
src_files = []

for file in input_files:
    src = rasterio.open(file)
    src_files.append(src)
    if smallest_raster_shape is None or (src.width * src.height) < (smallest_raster_shape[0] * smallest_raster_shape[1]):
        smallest_raster_shape = (src.height, src.width)
    data_type = src.meta['dtype']  # Get the data type from metadata

# Create a new array to hold the merged data
merged_arrays = []

# Read and resample data from each input dataset into separate arrays
for i, src in enumerate(src_files):
    src_resampled = np.empty((src.count, smallest_raster_shape[0], smallest_raster_shape[1]), dtype=data_type)
    src.read(out=src_resampled, resampling=Resampling.bilinear)
    merged_arrays.append(src_resampled)

# Combine the arrays to create the merged dataset
merged_array = np.concatenate(merged_arrays, axis=0)

# Update metadata
out_meta = src_files[0].meta.copy()
out_meta.update({
    "count": merged_array.shape[0],  # Set the number of bands
    "height": smallest_raster_shape[0],
    "width": smallest_raster_shape[1]
})

# Set band names in metadata
out_meta['dtype'] = 'float32'  # Update dtype if needed
out_meta['nodata'] = np.nan  # Set nodata value if needed
out_meta['count'] = merged_array.shape[0]  # Update band count
out_meta['descriptions'] = band_names

# Output file path for the merged GeoTIFF
output_file = r'C:\Users\bleser\OneDrive - Stichting Deltares\Desktop\Floods&Health Quantification\Data bronnen\tif_raw_moz\Output\merged_tif_data_NEW3.tif'  # Replace with your desired output path

# Write the merged dataset to a new GeoTIFF file
with rasterio.open(output_file, "w", **out_meta) as dest:
    dest.write(merged_array)

print("Merge completed successfully.")
```

```python
import numpy as np
import xarray as xr
import rasterio

# Load the population map TIFF file using rasterio
population_map_path = r'C:\FloodsAndHealthTool-main\examples\Data bronnen\Pop_dens_moz.tif'
with rasterio.open(population_map_path) as pop_src:
    population_map_array = pop_src.read(1)

# Load the age map TIFF file for children using rasterio
children_map_path = 'C:/Users/bleser/OneDrive - Stichting Deltares/Desktop/Floods&Health Quantification/Data bronnen/tif_raw_moz/Output/combined_output_raster_children.tif'
with rasterio.open(children_map_path) as children_src:
    children_map_array = children_src.read(1)

# Load the age map TIFF file for adults using rasterio
adults_map_path = 'C:/Users/bleser/OneDrive - Stichting Deltares/Desktop/Floods&Health Quantification/Data bronnen/tif_raw_moz/Output/combined_output_raster_adults.tif'
with rasterio.open(adults_map_path) as adults_src:
    adults_map_array = adults_src.read(1)

# Extract the region of interest (provide the coordinates accordingly)
# For example, specify the top-left and bottom-right coordinates of the region of interest.
top_left_x, top_left_y = 0.66e6, 7.84e6
bottom_right_x, bottom_right_y = 0.71e6, 7.84e6

# Get the pixel coordinates corresponding to the geographic coordinates
top_left_row, top_left_col = pop_src.index(top_left_x, top_left_y)
bottom_right_row, bottom_right_col = pop_src.index(bottom_right_x, bottom_right_y)

# Crop the arrays to the region of interest
population_map_roi = population_map_array[top_left_row:bottom_right_row, top_left_col:bottom_right_col]
children_map_roi = children_map_array[top_left_row:bottom_right_row, top_left_col:bottom_right_col]
adults_map_roi = adults_map_array[top_left_row:bottom_right_row, top_left_col:bottom_right_col]

# Calculate the number of children and adults in the region of interest
num_children_roi = children_map_roi.sum()
num_adults_roi = adults_map_roi.sum()

# Assuming you have the population density map as a separate TIFF file using rasterio
population_density_map_path =  r'C:\FloodsAndHealthTool-main\examples\Data bronnen\Pop_dens_moz.tif'
with rasterio.open(population_density_map_path) as pop_density_src:
    population_density_map_array = pop_density_src.read(1)

# Crop the population density map to the region of interest using rasterio
population_density_roi = population_density_map_array[top_left_row:bottom_right_row, top_left_col:bottom_right_col]

# Create xarray DataArrays for the data
population_density_da = xr.DataArray(population_density_roi, dims=('latitude', 'longitude'))
# Create one-element NumPy arrays to represent the number of children and adults
num_children_da = xr.DataArray(np.array([num_children_roi]), dims=('region_of_interest',))
num_adults_da = xr.DataArray(np.array([num_adults_roi]), dims=('region_of_interest',))

# Combine DataArrays into a Dataset
data_vars = {
    'population_density': population_density_da,
    'num_children': num_children_da,
    'num_adults': num_adults_da
}

# Define coordinates for the Dataset
latitude = np.linspace(-90, 90, population_map_roi.shape[0])
longitude = np.linspace(-180, 180, population_map_roi.shape[1])
region_of_interest = [f"Region of Interest {top_left_x}-{top_left_y}-{bottom_right_x}-{bottom_right_y}"]

# Create the Dataset
dataset = xr.Dataset(data_vars, coords={'latitude': latitude, 'longitude': longitude, 'region_of_interest': region_of_interest})

# Save the Dataset to a NetCDF file
output_file_netcdf = 'C:/Users/bleser/OneDrive - Stichting Deltares/Desktop/Floods&Health Quantification/Data bronnen/tif_raw_moz/combined_data.nc'
dataset.to_netcdf(output_file_netcdf)
print("Combined data saved to NetCDF file:", output_file_netcdf)
```
